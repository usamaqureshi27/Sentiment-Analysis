```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://konbriefing.com/en-topics/hacker-attacks-germany.html'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text , 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')
print(table)
```

```python
titles = soup.find_all('th')
titles
```

```python
table_titles = [title.text for title in titles]
table_titles
```

```python
df = pd.DataFrame(columns= table_titles)
df
```

```python
data = soup.find_all('tr')
```

```python
for row in data:
    row_data = row.find_all('td')
    table_data = [datas.text for datas in row_data]
    print(table_data)
```

```python
for row in data[1:]:
    row_data = row.find_all('td')
    table_data = [datas.text for datas in row_data]

    # df = pd.concat([df, table_data], ignore_index=True)
    
    # df = df.append(table_data, ignore_index=True)
    
    length = len(df)
    df.loc[length] = table_data
```
